# StandardScaler

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
# fit the scaler to the train set , it will learn the parameters
scaler.fit(X_train)
```

```python
# transform train and test set

X_train_scaler = scaler.transform(X_train)
X_test_scaler = scaler.transform(X_test)
```

# Effect of Scaling

```python
fig , (ax1, ax2) = plt.subplots(ncols = 2, figsize = (12,5))
ax1.scatter(X_train['Age'],X_train['EstimatedSalary'])
ax1.set_title('Before Scaling')

ax2.scatter(X_train_scaler['Age'], X_train_scaler['EstimatedSalary'],color = 'red' )
ax2.set_title('After scaling')

plt.show()
```

```python
fig, (ax1,ax2) = plt.subplots(ncols = 2, figsize = (12,5))

ax1.set_title('Before Scaling')

sns.kdeplot(X_train['Age'],ax =ax1)
sns.kdeplot(X_train['EstimatedSalary'], ax = ax1)

# After Scaling

ax2.set_title('After Scaling')

sns.kdeplot(X_train_scaler['Age'], ax = ax2)
sns.kdeplot(X_train_scaler['EstimatedSalary'], ax = ax2)
```

# Why is Scaling  important?

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
lr_scaled = LogisticRegression()
```

```python
lr.fit(X_train,y_train)
lr_scaled.fit(X_train_scaler,y_train)
```

```python
y_pred = lr.predict(X_test)
y_pred_scaled = lr_scaled.predict(X_test_scaler)
```

```python
from sklearn.metrics import accuracy_score
```

```python
print('Actual', accuracy_score(y_test,y_pred))
print('scaled', accuracy_score(y_test,y_pred_scaled))
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt = DecisionTreeClassifier()
```

```python
dt_scaled = DecisionTreeClassifier()
```

```python
dt.fit(X_train,y_train)
```

```python
dt_scaled.fit(X_train_scaler,y_train)
```

```python
y_pred = dt.predict(X_test)
y_pred_scaled = dt_scaled.predict(X_test_scaler)
```

```python
print('Actual', accuracy_score(y_test, y_pred))
print('Scaled', accuracy_score(y_test,y_pred_scaled))
```

```python
new_rows = pd.DataFrame({
    'Age': [5, 90, 95],
    'EstimatedSalary': [1000, 250000, 350000],
    'Purchased': [0, 1, 1]
})

df = pd.concat([df, new_rows], ignore_index=True)
```

```python
plt.scatter(df['Age'], df['EstimatedSalary'])
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    df.drop('Purchased', axis=1),
    df['Purchased'],
    test_size=0.3,
    random_state=0
)

X_train.shape, X_test.shape
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

# fit the scaler to the train set, it will learn the parameters
scaler.fit(X_train)

# transform train and test sets
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
X_train_scaled = pd.DataFrame(X_train_scaled, columns=X_train.columns)
X_test_scaled = pd.DataFrame(X_test_scaled, columns=X_test.columns)
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 5))

ax1.scatter(X_train['Age'], X_train['EstimatedSalary'])
ax1.set_title("Before Scaling")
ax2.scatter(X_train_scaled['Age'], X_train_scaled['EstimatedSalary'],color='red')
ax2.set_title("After Scaling")
```
