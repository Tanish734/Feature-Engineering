```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('covid_toy.csv')
```

```python
df.head()
```

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder
```

```python
df.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(df.drop(columns=['has_covid']), df['has_covid'],test_size = 0.2)
```

```python
y_train
```

```python
# First
```

```python
si = SimpleImputer()
```

```python
X_train_fever = si.fit_transform(X_train[['fever']])

X_test_fever = si.fit_transform(X_test[['fever']])

X_train_fever
```

```python
# OrdinalEncoder - > cough

oe = OrdinalEncoder(categories = [['Mild','Strong']])

X_train_cough = oe.fit_transform(X_train[['cough']])

X_train_cough.shape
```

```python
X_test_cough = oe.fit_transform(X_test[['cough']])
```

```python
X_test_cough
```

```python
# OneHotEncoding -> gender,city

ohe = OneHotEncoder(drop = 'first',sparse_output = False)

X_train_gender_city = ohe.fit_transform(X_train[['gender','city']])


X_test_gender_city = ohe.fit_transform(X_test[['gender','city']])
```

```python
X_train_gender_city
```

```python
# Extracting Age

X_train_age = X_train.drop(columns = ['gender','fever', 'cough','city']).values

#Also

X_test_age = X_test.drop(columns = ['gender','fever','cough','city'])

X_train_age.shape
```

```python
X_train_transform = np.concatenate((X_train_age,X_train_gender_city,X_train_cough,X_train_fever),axis =1)

# also
X_test_transform = np.concatenate((X_test_age,X_test_gender_city,X_test_cough,X_test_fever),axis =1)
```

```python
X_train_transform
```

```python
# second step

from sklearn.compose import ColumnTransformer
```

```python
transformer = ColumnTransformer(transformers = [
    ('tnf1',SimpleImputer(), ['fever']),
    ('tnf2',OrdinalEncoder(categories =[['Mild','Strong']]),['cough']),
    ('tnf3', OneHotEncoder(drop = 'first',sparse_output = False),['gender','city'])
])
```
